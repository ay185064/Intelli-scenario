```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import re

# data cleaning
actual_data = []
for line in text.split('\n'):
  line = line.strip().lower()

  # remove examples
  if('examples' in line or 'scenario' in line or '|' in line or '@' in line or len(line) == 0 or 'feature' in line or 'background' in line or 'product' in line or 'system' in line):
    continue

  # extract punc
  line = re.sub('\'.*\'','',line)
  line = re.sub('<.*>','',line)
  line = re.sub('[0-9]+','',line)
  line = re.sub('$','',line).strip()
  
  actual_data.append(line)
```

```python
size = 3
stopwords = set(['given','when','then','the','and'])
final_words = []
def stop_words():
  for line in actual_data:
    line = line.split()
    new_line = []

    for word in line:
      if word not in stopwords:
        new_line.append(word)
    new_line[-1] = new_line[-1]+"<end>"
    line_size = len(new_line)%size
    new_line += ['<pad>']*line_size
    print(new_line)
    final_words.extend([new_line[i:i+size] for i in range(len(new_line)-size)])        
stop_words()
```

```python
s = set()
uniq_words = []
for i in final_words:
  if tuple(i) not in s:
    s.add(tuple(i))
    uniq_words.append(i)
print(len(uniq_words))
final_words = uniq_words
```

```python
import tensorflow as tf
import numpy as np
```

```python
# encode sequence
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(final_words)
sequences = tokenizer.texts_to_sequences(final_words)
```

```python
seq_array = np.array(sequences)
vocab_size = len(tokenizer.word_index) + 1
```

```python
X, y = seq_array[:,:-1], seq_array[:,-1]
```

```python
y = tf.keras.utils.to_categorical(y, num_classes=vocab_size)
seq_length = X.shape[1]
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(vocab_size, 50, input_length=seq_length))
model.add(tf.keras.layers.LSTM(100, return_sequences=True))
model.add(tf.keras.layers.LSTM(100))
model.add(tf.keras.layers.Dense(100, activation='relu'))
model.add(tf.keras.layers.Dense(vocab_size, activation='softmax'))
print(model.summary())
```

```python
# compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit model
model.fit(X, y, batch_size=32, epochs=200)
```

```python
def generate_words(inp):
  orig_ip = inp
  result = []
    
  for _ in range(7):
    encode = tokenizer.texts_to_sequences([inp])[0]
    encode = tf.keras.preprocessing.sequence.pad_sequences([encode], maxlen=seq_length, truncating='pre')
    
    # predict probabilities for each word
    yhat = model.predict_classes(encode, verbose=0)[0]
    
    # map predicted word index to word
    out_word = ''
    for word, index in tokenizer.word_index.items():
      if index == yhat:
        out_word = word
        break
    
    inp += ' ' + out_word
    if('<end>' in out_word):
      result.append(out_word.replace('<end>',''))
      break
    if('<pad>' == out_word):
      break
    
    result.append(out_word)
    ' '.join(result)

  print("Input:  "+orig_ip)
  print("Output: "+orig_ip+" "+" ".join(result))
```
